# Retrieve Targets (Stage 0.2.0)

This notebook retrieves the targets for particular ligand
class, refer to `targets` (from `src.CONST.targets.Targets`), 
using Biopython `Bio.PDB.PDBList`'s `download_pdb_files` function.

## Preliminaries

```python
from Bio.PDB import PDBList

from src.CONST.targets import Targets
from src.utils.logger import Logger
from src.utils.housekeeping import (
    fix_dir,
    log_pkg_ver,
    remove_prev_data
)
```

```python
log = Logger("stage-0.2.0-retrieve_targets")
```

```python
pkg_imports: list[str] = ["biopython"]
log_pkg_ver(pkg_imports, log)
```

```python
DIR: list[str] = [
        "../data/targets"
    ]
```

```python
remove_prev_data(DIR, log)
```

```python
fix_dir(DIR, log)
```

## Target Retrieval

```python
targets = Targets.TARGETS
targets
```

```python
targets = list(
        targets.keys()
    )
targets
```

```python
for tar in targets:
    log.info(f"Retrieving {tar} from PDB ...")
    PDBList().retrieve_pdb_file(
        pdb_code=tar,
        file_format="pdb",
        overwrite=True,
        pdir="../data/targets"
    )
```
