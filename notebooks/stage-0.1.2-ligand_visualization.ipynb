```python
dataset = pd.read_csv(
        "../data/natural_antibiotics.csv", index_col=0
    )
```

```python
mirror_df = pd.read_csv("../data/mirror_ligand.csv", index_col=0)
```

```python
mirror_df
```

Group the antibiotics per mechanism/class, ie beta-lactams, tetracyclines, per row.

```python
native: list[Mol] = []
nlegend: list[str] = [] # native molecule
mirror: list[Mol] = [] # mirror-type molecule
mlegend: list[str] = []

for ab_arr in AB_ARR:
    submirror: list[Mol] = [] # mirror-type molecule
    sublegend_m: list[str] = []
    subnative: list[Mol] = [] # native molecule
    sublegend_n: list[str] = []

    for ab in ab_arr:
        ab = ab.title()
        cid = dataset[ab]["cid"]

        mirror_mol: Mol = Chem.MolFromSmiles(
                mirror_df[ab]["mirror_smi"]
            )
        native_mol: Mol = Chem.MolFromSmiles(
                dataset[ab]["smiles"]
            )

        submirror.append(mirror_mol) # mirror-type molecule
        sublegend_m.append(f"Mirror {ab.title()}")
        subnative.append(native_mol) # native molecule
        sublegend_n.append(ab.title())

    mirror.append(submirror) # mirror-type molecule
    mlegend.append(sublegend_m)
    native.append(subnative) # native molecule
    nlegend.append(sublegend_n)
```

## Native Ligands

```python
img = Draw.MolsMatrixToGridImage(
    molsMatrix=native,
    legendsMatrix=nlegend,
    subImgSize=(512, 512),
    useSVG=True,
    maxMols=max(
        [len(n) for n in nlegend]
    ) * len(nlegend) * 2

)

with open(
        "../output/figs/ligands/ligands-X.svg", "w", encoding="utf-8"
    ) as out:
    out.write(img.data)
img
```

## Mirror Ligands

```python
img = Draw.MolsMatrixToGridImage(
    molsMatrix=mirror,
    legendsMatrix=mlegend,
    subImgSize=(512, 512),
    useSVG=True,
    maxMols=max(
            [len(n) for n in mlegend]
        ) * len(mlegend) * 2
)

with open(
        "../output/figs/ligands/mirror_ligands-X.svg", "w", encoding="utf-8"
    ) as out:
    out.write(img.data)
img
```
