## Preliminaries

```python
import pandas as pd
from pubchempy import Compound, get_compounds, download

from src.CONST.ligands import Ligands
```

## Dataset

```python
antibiotics = Ligands.ANTIBIOTICS
antibiotics
```

```python
ab_cid: dict[str, dict] = {}

for ab in antibiotics:
    res = get_compounds(ab, "name", record_type="3d")
    if len(res) == 0:
        res = get_compounds(ab, "name")
    cid = res[0].cid
    ab_cid[ab] = Compound.from_cid(cid).to_dict(
            [
                "cid",
                "smiles",
                "inchi",
                "coordinate_type",
                "xlogp",
                "tpsa",
                "h_bond_donor_count",
                "h_bond_acceptor_count",
                "rotatable_bond_count",
                "cactvs_fingerprint"
            ]
        )
    download(
        "sdf", f"../data/ligands/sdf/{cid}.sdf", cid, overwrite=True
    )
```

```python
compound_query = pd.DataFrame(ab_cid)
compound_query
```

```python
compound_query.to_csv("../data/natural_antibiotics.csv")
```
