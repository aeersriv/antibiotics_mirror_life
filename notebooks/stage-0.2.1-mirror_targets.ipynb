Generate the enantiomer of targets by manually inverting each

## Preliminaries


```python
from glob import iglob

from src.utils.generate_mirror import invert_pdb
from src.CONST.targets import Targets
```

## Generate Mirror Targets

```python
targets = Targets.TARGETS
targets
```

```python
targets = [tar.lower() for tar in targets.keys()]
targets
```

```python
dir: str = "../data/targets/*"

for file in iglob(dir):
    filename: str = file.split("/")[-1]

    if filename[:4] in targets:
        print(filename)
        invert_pdb(
            file, f"../data/mirror_targets/"
        )
```
