## Generate Mirror Ligand

```python
dataset = pd.read_csv(
        "../data/antibiotics_pubchem_query.csv", index_col=0
    )
```

```python
dataset.head()
```

```python
mirror_smi_arr: list[str] = []

for ab in dataset:
    ref_smi = dataset[ab]["smiles"]
    mol = Chem.MolFromSmiles(dataset[ab]["smiles"])

    for atom in mol.GetAtoms():
        atom.InvertChirality()

    mirror_smi_arr.append(
        mirr_smi := Chem.MolToSmiles(mol)
    )
    print(
        f"Native: {ref_smi}\n"
        + f"Mirror: {mirr_smi}\n"
    )
```

```python
mirror_df = dataset.copy(deep=True)
mirror_df = mirror_df.drop(
    [
        "xlogp",
        "tpsa",
        "h_bond_donor_count",
        "h_bond_acceptor_count",
        "rotatable_bond_count",
        "cactvs_fingerprint"
    ]
)
```

```python
mirror_df.loc["mirror_smi"] = mirror_smi_arr
```
