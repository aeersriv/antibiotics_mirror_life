## Preliminaries

```python
import matplotlib_inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
matplotlib_inline.backend_inline.set_matplotlib_formats('retina')
```

```python
okabe_ito_cs_arr: list[str] = [
	"#E69F00",
	"#56B4E9",
	"#009E73",
	"#F0E442",
	"#0072B2",
	"#D55E00",
	"#CC79A7",
	"#000000"
]
```

```python
okabe_ito_palette = sns.color_palette(
	okabe_ito_cs_arr,
	len(okabe_ito_cs_arr)
)
print(okabe_ito_palette.as_hex())
```

## Dataset

```python
antibiotics = pd.read_csv("../data/ANTIBIOTICS-FLD1.csv")
antibiotics
```

```python
ab_classes = antibiotics["antibiotic_class"].value_counts()
who_classes = antibiotics["who_class"].value_counts()
ab_classes, who_classes
```

```python
ac_labels: list[str] = [
        "Beta-lactams",
        "Tetracyclines",
        "Aminoglycosides",
        "Macrolides",
        "Quinolones",
        "Glycopeptides",
        "Glycylcyclines",
        "Amphenicols",
        "Oxazolidinones",
        "Ansamycin",
        "Phosphonic Acids",
        "Trimethoprims",
        "Sulfonamides"
    ]

wc_labels: list[str] = [
        "Critically Important",
        "Highly Important"
    ]
```

```python
plt.clf()
fig, ax = plt.subplots(figsize=(4, 4))
ax.pie(
    ab_classes,
    colors=okabe_ito_palette,
    labels=ac_labels,
    radius=3,
    wedgeprops={"linewidth": 1}
)
plt.axis("off")
plt.savefig("../output/antibiotic_classes.svg")
plt.show()
```

```python
plt.clf()
fig, ax = plt.subplots(figsize=(4, 4))

ax.pie(
    who_classes,
    labels=wc_labels,
    colors=okabe_ito_palette,
    radius=3,
    wedgeprops={"linewidth": 1}
)
plt.axis("off")
plt.savefig("../output/who_classes.svg")
plt.show()
```
