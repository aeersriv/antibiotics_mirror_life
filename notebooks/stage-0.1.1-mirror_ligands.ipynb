## Generate Mirror Ligand

```python
dataset = pd.read_csv(
        "../data/natural_antibiotics.csv", index_col=0
    )
```

```python
dataset.head()
```

```python
mirror_smi_arr: list[str] = []

for ab in dataset:
    ref_smi = dataset[ab]["smiles"]
    cid = dataset[ab]["cid"]
    mirror_mol = Chem.MolFromSmiles(dataset[ab]["smiles"])

    for atom in mirror_mol.GetAtoms():
        atom.InvertChirality()

    mirror_smi_arr.append(
        mirr_smi := Chem.MolToSmiles(mirror_mol)
    )
    Chem.MolToMolFile(
        mirror_mol, f"../data/mirror_ligands/mol2/mirror-{cid}.mol"
    )
    log.info(
        f"Successfully generated enantiomer of {ab}:\n"
        + f"\tNative: {ref_smi}\n\tMirror: {mirr_smi}"
    )
```

```python
mirror_df = dataset.copy(deep=True)
```

```python
mirror_df.loc["mirror_smi"] = mirror_smi_arr
```
